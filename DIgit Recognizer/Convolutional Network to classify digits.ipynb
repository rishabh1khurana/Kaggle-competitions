```python
###Model 2####
model2 = Sequential()


####
# Add the first hidden layer
model2.add(Conv2D(32,(3,3),activation='relu',input_shape=(28,28,1)))
model2.add(MaxPool2D(pool_size=(2,2)))
model2.add(Conv2D(64,(3,3),activation='relu'))
model2.add(MaxPool2D(pool_size=(2,2)))
model2.add(Conv2D(128,(3,3),activation='relu'))
model2.add(MaxPool2D(pool_size=(2,2)))
model2.add(Flatten())

model2.add(Dense(512, activation='relu', input_shape=(784,)))

# Add the second hidden layer
model2.add(Dense(256, activation='relu'))

#model2.add(Dense(128, activation='relu'))
#model2.add(Dense(32, activation='relu'))
# Add the output layer
model2.add(Dense(10, activation='softmax'))

# Compile the model with rmsprop optimizer
model2.compile(optimizer='rmsprop',
             loss='categorical_crossentropy',
            metrics=['accuracy'])
    
  #  also try 
    #model2.compile(optimizer='SGD(lr=.01,momentum=.3)',
    #         loss='categorical_crossentropy', metrics=['accuracy'])
    
early_stopping_monitor = EarlyStopping(patience=3)
model2.fit(train,target,validation_split=0.3,batch_size=128, epochs=40, callbacks=[early_stopping_monitor])
#model2.save("model_3.h5")


```

```python
#### Try using Rmsprop or other optimizer in this model

model = Sequential()

model.add(Conv2D(filters = 32, kernel_size = (5,5),padding = 'Same', 
#                 activation ='relu', input_shape = (28,28,1)))
model.add(Conv2D(filters = 32, kernel_size = (5,5),padding = 'Same', 
#                 activation ='relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.25))


model.add(Conv2D(filters = 64, kernel_size = (3,3),padding = 'Same', 
                 activation ='relu'))
model.add(Conv2D(filters = 64, kernel_size = (3,3),padding = 'Same', 
                 activation ='relu'))
model.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))
model.add(Dropout(0.25))


model.add(Flatten())
model.add(Dense(256, activation = "relu"))
model.add(Dropout(0.5))
model.add(Dense(10, activation = "softmax"))


optimizer = RMSprop(lr=0.001, rho=0.9, epsilon=1e-08, decay=0.0)
model.compile(optimizer = optimizer , loss = "categorical_crossentropy", metrics=["accuracy"])

early_stopping_monitor = EarlyStopping(patience=3)
history = model.fit(train,target, validation_split=0.3, batch_size = 16, epochs = 30, callbacks=[early_stopping_monitor])

```
